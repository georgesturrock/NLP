#### Homework 3 - Question 1a
Question 1a asks us to calculate the edit distance between my given first name and nickname.  The edit distance is a method used to quantify the similarity of two strings.  The edit distance calculates similarity by measuring the number of changes, or edits, needed to transform one string to another.  In this instance, we will calculate the number of edits required to transform my given first name to my nickname.  

With a one syllable first name, nicknames were fairly rare for me.  One that sticks out is "Pappageorgio".  A friend coined this name after a character in National Lampoon's Vegas Vacation.  There are obvious phonetic similarities between the two names, but what is the textual similarity as calculated by the edit distance method?  The charts shown below show the matrix and straight line methods of manually calculating the edit distance.  The highlighted green cells show the actual edit distance between "george" and "pappageorgio".  

The matrix method is a rather complicated method which tracks the number of edit required to change the columnwise of the given name to the columnwise values of the nickname.  For instance, the number of edits requried to convert "g" to "pap" is three.  One edit would be to change "g" to "p".  The second edit is to add "a".  The third edit is to add "p".  The straight line method simply tracks the columnwise changes between the two names.  Both result in the same edit distance of seven.

```python
display(Image.open('pappageorgioED.png'))
```

NLTK also provides a function to calculate the edit distance between the name and nickname.  This function confirms the manual calculations above by producing an edit distance of seven.

```python
name = 'george'
nickname2 = 'pappageorgio'
ed = edit_distance(name, nickname2)
ed
```

#### Homework 3 - Question 1b
Question 1b from Homework 3 inquires about the percentage string match between my given name and nickname.  This is a simple calculation where number of characters in common between the two names is divided by the total characters of the longest string.  The edit distance calculated above can assist with this calculation.  Subtracting the edit distance from the longest string returns the number of characters in common between the two strings.  The function below calculates the percent string match.

```python
def pctStringMatch(one, two, edist):
    if len(one) != 0 and len(two) != 0:
        long = (max([one, two], key=len))
        out = (len(long)-edist)/(len(long))
        return out
    else:
        return 0
```

The calculated percent string match in shown below:

```python
pct = pctStringMatch(name, nickname2, ed)
round(pct, 4)
```

#### Homework 3 - Question 2
In 3.2, the first two sentences were taken from a recent book my wife read.  The book is "The Night Circus" by Erin Morgenstern.  The first two sentences are "The circus arrives with warning.  No accnouncements precede it, no paper notices on downtown posts and billboards, no mentions or advertisements in local newspapers."  The stop words were removed from these sentences.  The remaining words were read back to my wife to see if she could identify the book.  

```python
sentences = "The circus arrives without warning. No announcements precede it, no paper notices on downtown posts and billboards, no mentions or advertisements in local newspapers."
sentences = sentences.lower()
table = str.maketrans('', '', string.punctuation)
sentences2 = nltk.word_tokenize(str(sentences))
sentences3 = [w.translate(table) for w in sentences2]
sentences4 = [w for w in sentences3 if w not in ensw and w.isalpha()]
```

After removing punctuation and english stop words, the remaining tokens from the first two sentences of "The Night Circus" are shown below.

```python
sentences4
```

My wife as able to correctly guess the title of the book on the first try.  This was a bit of a foregone conclusion as one of the three words in the book's title appears in the first two sentences of the book.  However, the remaining words in the "sentences4" variable are somewhat readable even with the stop words removed.  The reader can get the overall impression of the first two sentences of the book by making reasonable assumptions on the missing words used to connect the remaining words.
